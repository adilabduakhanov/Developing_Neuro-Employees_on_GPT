```python
from google.colab import auth
auth.authenticate_user()

import gspread
from google.auth import default

creds, _ = default()
gc = gspread.authorize(creds)
```

```python
spreadsheet = gc.open("ыфваапНовая таблица")
worksheet = spreadsheet.sheet1
```

```python
headers = [
    "Имя",
    "Контакт",
    "Email",
    "Адрес",
    "Дата замера",
    "Примечания",
    "Дата записи",
    "Услуга",
    "Тип объекта",
    "Предпочтения",
    "Доп услуги"
]

worksheet.clear()
worksheet.append_row(headers)
```

```python
scenarios = [
    "Подбор и установка новых окон",
    "Свой индивидуальный вариант",
    "Консультация с менеджером компании"
]
```

```python
from datetime import datetime

def save_to_sheet(data):
    worksheet.append_row(data)
```

```python
def scenario_windows():
    name = input("Имя: ")
    contact = input("Контакт: ")
    email = input("Email: ")
    address = input("Адрес: ")
    measure_date = input("Дата замера: ")
    notes = input("Примечания: ")
    object_type = input("Тип объекта: ")
    preferences = input("Предпочтения: ")
    extra_services = input("Доп услуги: ")

    service = "Подбор и установка новых окон"
    record_date = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    data = [
        name,
        contact,
        email,
        address,
        measure_date,
        notes,
        record_date,
        service,
        object_type,
        preferences,
        extra_services
    ]

    save_to_sheet(data)
```

```python
def scenario_custom():
    name = input("Имя: ")
    contact = input("Контакт: ")
    email = input("Email: ")

    service = "Свой индивидуальный вариант"
    record_date = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    data = [
        name,
        contact,
        email,
        "",   # Адрес
        "",   # Дата замера
        "",   # Примечания
        record_date,
        service,
        "",   # Тип объекта
        "",   # Предпочтения
        ""    # Доп услуги
    ]

    save_to_sheet(data)
```

```python
def scenario_consultation():
    name = input("Имя: ")
    contact = input("Контакт: ")
    email = input("Email: ")

    service = "Консультация с менеджером компании"
    record_date = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    data = [
        name,
        contact,
        email,
        "",
        "",
        "",
        record_date,
        service,
        "",
        "",
        ""
    ]

    save_to_sheet(data)
```

```python
print("Выберите сценарий:")
print("1 - Подбор и установка новых окон")
print("2 - Свой индивидуальный вариант")
print("3 - Консультация с менеджером компании")

choice = input("Введите номер: ")

if choice == "1":
    scenario_windows()
elif choice == "2":
    scenario_custom()
elif choice == "3":
    scenario_consultation()
else:
    print("Неверный выбор")
```
