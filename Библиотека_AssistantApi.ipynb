```python
from openai import OpenAI
from google.colab import userdata

api_key = userdata.get("OPENAI_API_KEY")

client = OpenAI(api_key=api_key)
```

```python
def focus_mentor(tasks, deadlines, energy_level, stress_level):

    prompt = f"""
    Ты интеллектуальный ассистент по продуктивности FocusMentor.

    Пользователь предоставил:
    Задачи: {tasks}
    Дедлайны: {deadlines}
    Уровень энергии: {energy_level}/10
    Уровень стресса: {stress_level}/10

    Выполни анализ:
    1. Определи приоритетность задач
    2. Оцени риск перегрузки
    3. Составь план на день
    4. Дай рекомендации по управлению энергией
    5. Добавь мотивационное сообщение

    Ответ дай структурированно.
    """

    response = client.responses.create(
        model="gpt-4.1-mini",
        input=prompt
    )

    return response.output[0].content[0].text
```

```python
tasks = [
    "Подготовить коммерческое предложение",
    "Провести звонок с клиентом",
    "Подготовить отчет по продажам",
    "Запустить рекламную кампанию"
]

deadlines = [
    "Сегодня 18:00",
    "Сегодня 15:00",
    "Завтра",
    "Через 3 дня"
]

energy = 6
stress = 7

result = focus_mentor(tasks, deadlines, energy, stress)

print(result)
```

Описание разработанного ассистента

Был разработан интеллектуальный агент FocusMentor — ассистент по продуктивности и управлению фокусом.

Цель:

Помогать пользователю планировать задачи, оценивать риски перегрузки и повышать эффективность работы.

Поведение:

Ассистент получает на вход:

список задач

дедлайны

уровень энергии

уровень стресса

На основе этих данных он:

расставляет приоритеты

анализирует риск выгорания

формирует план действий

дает рекомендации

генерирует мотивационное сообщение

Используемые инструменты:

OpenAI Responses API

Модель gpt-4.1-mini

Результат:

Пользователь получает структурированный аналитический отчет с рекомендациями и планом действий.
