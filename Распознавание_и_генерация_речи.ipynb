```python
from openai import OpenAI
from google.colab import userdata
import os
from IPython.display import Audio
from jiwer import wer

# ключ
key = userdata.get("OPENAI_API_KEY")
os.environ["OPENAI_API_KEY"] = key

client = OpenAI()
```

```python
poem_response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "system", "content": "Ты поэт."},
        {"role": "user", "content": "Напиши короткое философское стихотворение о будущем технологий."}
    ],
    temperature=0.9
)

poem_text = poem_response.choices[0].message.content

print("Сгенерированное стихотворение:\n")
print(poem_text)
```

```python
speech_file_path = "poem_audio.mp3"

with client.audio.speech.with_streaming_response.create(
    model="gpt-4o-mini-tts",
    voice="alloy",
    input=poem_text
) as response:
    response.stream_to_file(speech_file_path)

print("Аудиофайл сохранён.")

Audio(speech_file_path)
```

```python
audio_file = open("poem_audio.mp3", "rb")

transcription = client.audio.transcriptions.create(
    model="gpt-4o-mini-transcribe",
    file=audio_file
)

recognized_text = transcription.text

print("Распознанный текст:\n")
print(recognized_text)
```

```python
error_rate = wer(poem_text, recognized_text)

print("\nWER:", error_rate)
```
