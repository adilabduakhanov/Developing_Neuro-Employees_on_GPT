```python
video_url = "ВСТАВЬТЕ_ССЫЛКУ_НА_ВИДЕО"

def download_audio(url, filename):
    ydl_opts = {
        'format': 'bestaudio/best',
        'outtmpl': filename,
        'postprocessors': [{
            'key': 'FFmpegExtractAudio',
            'preferredcodec': 'mp3',
        }],
    }

    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        ydl.download([url])

download_audio(video_url, "/content/video_audio")
```
