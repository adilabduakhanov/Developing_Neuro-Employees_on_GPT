Задание 3

```python
data = {
    'Наименование товара': ['Телевизор', 'Ноутбук', 'Смартфон', 'Наушники', 'Планшет'],
    'Цена': [50000, 75000, 30000, 5000, 20000],
    'Остаток': [10, 5, 25, 50, 15]
}

df3 = pd.DataFrame(data)

new_row = {'Наименование товара': 'Умная колонка', 'Цена': 8000, 'Остаток': 30}
df3 = pd.concat([df3, pd.DataFrame([new_row])], ignore_index=True)

display(df3)
```

Задание 4

```python
df3_sorted = df3.sort_values(by='Цена', ascending=True).reset_index(drop=True)
display(df3_sorted)
```

Задание 5

```python
df3_sorted.to_excel('sorted_products.xlsx', index=False)
print("DataFrame 'df3_sorted' saved to 'sorted_products.xlsx'")
```
