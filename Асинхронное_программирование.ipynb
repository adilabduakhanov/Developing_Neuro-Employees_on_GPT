```python
import aiohttp
import asyncio

async def get_weather(city: str) -> None:
    url = "https://api.openweathermap.org/data/2.5/weather"

    params = {
        "q": city,
        "appid": "90bda1af1e49a3fb45134524de194dd0",  # —Ç–≤–æ–π –∫–ª—é—á
        "units": "metric",
        "lang": "ru"
    }

    async with aiohttp.ClientSession() as session:
        async with session.get(url, params=params) as response:
            data = await response.json()

            if response.status != 200:
                print(f"–û—à–∏–±–∫–∞ –¥–ª—è –≥–æ—Ä–æ–¥–∞ {city}: {data.get('message')}")
                return

            print(f"\n–ü–æ–≥–æ–¥–∞ –≤ {city}:\n")
            print(f"–¢–µ–º–ø–µ—Ä–∞—Ç—É—Ä–∞: {data['main']['temp']}¬∞C")
            print(f"–û–ø–∏—Å–∞–Ω–∏–µ: {data['weather'][0]['description'].capitalize()}")
            print(f"–í–µ—Ç–µ—Ä: {data['wind']['speed']} –º/—Å")
            print(f"–í–ª–∞–∂–Ω–æ—Å—Ç—å: {data['main']['humidity']}%")
            print(f"–î–∞–≤–ª–µ–Ω–∏–µ: {data['main']['pressure']} –≥–ü–∞")


async def main():
    cities = ["Moscow", "London", "Paris"]
    tasks = [get_weather(city) for city in cities]
    await asyncio.gather(*tasks)

# üîπ –í Colab –Ω—É–∂–Ω–æ —Ç–∞–∫:
await main()
```
